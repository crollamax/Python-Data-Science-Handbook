# Scatter Plots

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## Scatter plots with plt.plot

The same function as in prev chapter (plt.plot()) can product scatter plots:

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

 The third argument in the function call represents the type of symbol to be used. Most of them are below:

```python
rng = np.random.RandomState(0)
plt.figure(figsize=(15,10))
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd', 'p']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label=f"""marker='{marker}'""")
plt.legend(numpoints=1, prop={'size':15})
plt.xlim(0, 1);
```

These can be combined with line and colour codes to plot points along with a connecting line:

```python
plt.plot(x, y, '-ok');
```

Additional keyword arguments allow for more control:

```python
plt.plot(x, y, '-p', color='gray',
        markersize=15, linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

## Scatter plots with plt.scatter

```python
plt.scatter(x, y, marker='o');
```

The main difference between plt.scatter and plt.plot is that it can create scatter plots where the properties of each individual point (size, face, color, edge color etc) can be individually controlled or mapped to data.

Consider a randomscatter plot with points of many colors and sizes. We can adjust the alpha (transparency) keyword to better see overlapping results:

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
           cmap='viridis')
plt.colorbar();
```

The color argument is automatically mapped to a color scale and the size argument is given in pixels. Thus, sthe color and size of points can be used to convey information.

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.3,
           s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

Thus we have explored four different dimensions of the data: the (x,y) corresponds to the sepal length and width, the size of the point relates to petal width, and the color is related to the particular species.

## plot vs scatter: Efficiency

When datasets become larger than a few thousand points, plt.plot can be noticeably more efficient that plt.scatter. The reason is that plt.scatter has to ability to render a different size and/or colour for each point, and so has to construct each point individually. In plt.plot, howover, all the points are essentially clones of each other, so the work in determining what the points look like need only be done once. Thus over thousands of points, this difference can be significant.
