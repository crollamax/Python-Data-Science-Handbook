# Multiple Subplots

Subplots are groups of smaller axes that can exist wtihin a single figure.

```python
%matplotlib inline 
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

##Â plt.axes: Subplots by Hand

The plt.axes function creates a standard axes object that fills the entire figure. plt.axes also takes an optional argument that is a list of four numbers [left, bottom, width, height] with ranges from 0 at the bottom left and 1 at the top right.

We can inset an exes at the top-right corner of another axes by setting the x and y position to 0.65 (ie starting at 65% of the width and 65% of the height of the figure) and the x and y extents to 0.2 (that is, the size of the axes is 20% of the width and 20% of the height):

```python
ax1 = plt.axes()  # standard axes
ax2 = plt.axes([0.65, 0.65, 0.2, 0.2])
```

The equivalent is fig.add_axes(). Lets add two vertically stacked axes:

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.5, 0.8, 0.4], 
                   xticklabels=[], ylim=(-1.2, 1.2))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.4],
                   ylim=(-1.2, 1.2))
x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

##Â plt.subplot: Simple Grid of Subplots

Aligned columns or rows of subplots are a common-enough need that there are lots of ways to do this. The lowest level is plt.subplot(), which creates a single subplot within a grid. As you can see. this takes three integer args - num_rows, num_cols, and index of plot to be created in the scheme, which runs from upper left to bottom right:

```python
for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
             fontsize=18, ha='center')
```

plt.subplots_adjust can adjust the spacing between these plots. The following uses the equvalent OO command:

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
            fontsize=18, ha='center')
```

hspace and wspace specifiy the spacing along the height and width of the figure, in units of subplot size i.e. 40%.

## plt.subplots: The Whole Grid in One Go

This can be tedious, when creating a large grid, especially if we want to hide the x- and y-axis labels on the inner plots. For this purpose, plt.subplots() is easier. This creates a fullgrid of subplots in a single line, returning them in a NumPy array. The arguments are the number of rows and number of columns, along with optional keywords sharex and sharey, which allows to specify the relationships between different axes.

```python
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

```python
# axes are in a two-dimensional array, indexed by [row, col]
for i in range(2):
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)),
                      fontsize=18, ha='center')
fig
```

##Â plt.GridSpec: More Complicated Arrangements

plt.GridSpec does not create a plot by itself, it's just a convenient interface that is recognised by the plt.subplot() command. For example, a gridsec for a grid of two rows and three columns with some specified height and width space looks like:

```python
grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)
```

```python
plt.subplot(grid[0, 0])
plt.subplot(grid[0, 1:])
plt.subplot(grid[1, :2])
plt.subplot(grid[1, 2]);
```

A good example of this: 

```python
# normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T

# set up axes with gridspec
fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on main axis
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)

# histogram on attached axes
x_hist.hist(x, 40, histtype='stepfilled',
            orientation='vertical', color='gray')
x_hist.invert_yaxis()

y_hist.hist(x, 40, histtype='stepfilled',
            orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```
